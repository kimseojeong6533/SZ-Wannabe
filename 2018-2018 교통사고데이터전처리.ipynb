```python
import csv
import pandas as pd
import numpy as np
import copy
f = pd.read_csv('TrafficAccident_copy1.csv')
f_copy=copy.deepcopy(f)
f_copy=pd.DataFrame()
f_copy
f.head(5)
print(f.shape)
```

# 데이터 전처리

```python
# '사고일시' -> 년/월/일/시로 바꾸기

year=[]
month=[]
day=[]
time=[]



for i in f['사고일시']:
    i=i.split()
    year.append(i[0][:-1])
    month.append(i[1][:-1])
    day.append(i[2][:-1])
    time.append(i[3])

    
f_copy['사고년도']=year
f_copy['사고월']=month
f_copy['사고일']=day
f_copy['사고시간']=time
f_copy.head(5)
```

```python
# 요일 -> 그대로 하되, 요일이라는 단어 빼고 월(0) ~ 일(6)
d1=f['요일'] 

sub1=[]
sub2=[]
for i in d1:
    sub2.append(i[0])
    
f_copy['사고요일']=sub2



for i in d1: 
    i=i[0]
    if i=='월':
        sub1.append(0)
    elif i=='화':
        sub1.append(1)
    elif i=='수':
        sub1.append(2)
    elif i=='목':
        sub1.append(3)
    elif i=='금':
        sub1.append(4)
    elif i=='토':
        sub1.append(5)
    else:   # 일 
        sub1.append(6)
    
    #sub1.append(i)

f_copy['사고요일_i']=sub1

# 시군구+도로명 -> 사고지 주소
Place=[]

for i in range(f.shape[0]):
    sub=''
    sub+=f['시군구'][i]+' '+f['도로명'][i]
    Place.append(sub)
f_copy['사고장소']=Place
f_copy

# 사고내용 -> 사고내용 그대로   # 부상신고(0), 경상(1), 중상(2), 사망(3)
c1=f['사고내용']

c2=[]
c=[]
for i in c1:
    c.append(i[:-2])

f_copy['사고내용']=c
    
for i in c1:
    if i[:-2]=='부상신고':
        c2.append(0)
    elif i[:-2]=='경상':
        c2.append(1)
    elif i[:-2]=='중상':
        c2.append(2)
    else:   # 사망
        c2.append(3)
        
    #c2.append(i[:-2])

f_copy['사고내용_i']=c2
f_copy.head(5)
```

```python
# 사고유형 -> 차대차(0)/ 차대사람(1)만

Type=[]
Type2=[]
    

for i in f['사고유형']:
    i=i.split()
    Type.append(i[0])
    if i[0]=='차대차':
        Type2.append(0)
    elif i[0]=='차대사람':
        Type2.append(1)
        
    #Type.append(i[0])
    
f_copy['사고유형']=Type
f_copy['사고유형_i']=Type2
f_copy.head(5)

# 기상상태 -> 그대로  / 맑음(0), 비(1), 흐림(2), 눈(3)

weather=f['기상상태']
w=[]
for i in weather:
    if i=='맑음':
        w.append(0)
    elif i=='비':
        w.append(1)
    elif i=='흐림':
        w.append(2)
    elif i=='눈':
        w.append(3)
        
        
f_copy['기상상태']=f['기상상태']    
f_copy['기상상태_i']=w
f_copy.head(5)
```

```python
# 가해운전자 차종, 가해운전자 성별, 피해운전자 차종, 피해운전자 성별 -> 그대로

f_copy['가해운전자 차종']=f['가해운전자 차종']
f_copy['가해운전자 성별']=f['가해운전자 성별']
f_copy['피해운전자 차종']=f['피해운전자 차종']
f_copy['피해운전자 성별']=f['피해운전자 성별']
f_copy.head(5)
```

```python
A_type=f['가해운전자 차종']  # 승용 : 0, 건설기계 : 1, 개인형이동수단(PM) : 2, 승합 : 3, 원동기 : 4, 이륜 : 5, 자전거 : 6, 특수 : 7, 화물 : 8
A_age=f['가해운전자 성별']
V_type=f['피해운전자 차종']
V_age=f['피해운전자 성별']
    
# 인덱싱 하기

# 가해운전자 차종
for i in range(len(A_type)):
    if A_type[i] == '승용':
        A_type[i]=0
    elif A_type[i]=='건설기계':
        A_type[i]=1
    elif A_type[i]=='개인형이동수단(PM)':
        A_type[i]=2
    elif A_type[i]=='승합':
        A_type[i]=3
    elif A_type[i]=='원동기':
        A_type[i]=4
    elif A_type[i]=='이륜':
        A_type[i]=5
    elif A_type[i]=='자전거':
        A_type[i]=6
    elif A_type[i]=='특수':
        A_type[i]=7
    elif A_type[i]=='화물':
        A_type[i]=8
        
f_copy['가해운전자 차종_i']=A_type

# 피해운전자 차종 # + 보행자, 사륜오토바이(ATV)
for i in range(len(V_type)):
    if V_type[i] == '승용':
        V_type[i]=0
    elif V_type[i]=='건설기계':
        V_type[i]=1
    elif V_type[i]=='개인형이동수단(PM)':
        V_type[i]=2
    elif V_type[i]=='승합':
        V_type[i]=3
    elif V_type[i]=='원동기':
        V_type[i]=4
    elif V_type[i]=='이륜':
        V_type[i]=5
    elif V_type[i]=='자전거':
        V_type[i]=6
    elif V_type[i]=='특수':
        V_type[i]=7
    elif V_type[i]=='화물':
        V_type[i]=8
    elif V_type[i]=='보행자':
        V_type[i]=9
    elif V_type[i]=='사륜오토바이(ATV)':
        V_type[i]=10
    
        
f_copy['피해운전자 차종_i']=V_type

# 가해운전자 성별
for i in range(len(A_age)):
    if A_age[i]=='남':
        A_age[i]=0
    else:
        A_age[i]=1
f_copy['가해운전자 성별_i']=A_age

# 피해운전자 성별
for i in range(len(V_age)):
    if V_age[i]=='남':
        V_age[i]=0
    else:
        V_age[i]=1
f_copy['피해운전자 성별_i']=V_age

f_copy.head(5)
```

```python
# 가해운전자 연령, 피해운전자 연령 -> 가해운전자 연령대, 피해운전자 연령대

import re
p=re.compile('[0-9]+')

a=f['가해운전자 연령']
b=f['피해운전자 연령']

a_1=[]
b_1=[]
for i in a:
    k=p.match(i)
    i=int(i[:k.end()])
    if i < 20:
        i=10
    elif i<30:
        i=20
    elif i<40:
        i=30
    elif i<50:
        i=40
    elif i<60:
        i=50
    elif i<70:
        i=60
    elif i<80:
        i=70
    elif i<90:
        i=80
    elif i<100:
        i=90
    a_1.append(i)

    
for i in b:
    k=p.match(i)
    i=int(i[:k.end()])
    if i < 20:
        i=10
    elif i<30:
        i=20
    elif i<40:
        i=30
    elif i<50:
        i=40
    elif i<60:
        i=50
    elif i<70:
        i=60
    elif i<80:
        i=70
    elif i<90:
        i=80
    elif i<100:
        i=90
    b_1.append(i)

# 연령대는 int타입
f_copy['피해운전자_연령대']=a_1
f_copy['가해운전자_연령대']=b_1
f_copy.head(5)

#f_copy=f_copy[['사고년도','사고월','사고일','사고시각','사고시간_인덱스','사고요일','사고장소','사고내용','사고유형','기상상태','가해운전자_차종','가해운전자_성별','가해운전자_연령대','피해운전자_차종','피해운전자_성별','피해운전자_연령대']]
#f_copy.head(10)
```

```python
# Time에서 3시간단위로 구분

'''
00~03 : 0
03-06 : 1
06-09 : 2
09-12 : 3
12-15 : 4
15-18 : 5
18-21 : 6
21-00 : 7

'''
t=f_copy['사고시간']

t_1=[]
t_origin=[]
# ~시 -> 숫자만 추출하여 int로 저장
for i in t:
    i=i[:2]
    t_1.append(int(i))

f_copy['사고시각']=t_1
sub=[]
for i in t_1:
    if 0<=i<3:
        i=0
    elif 3<=i<6:
        i=1
    elif 6<=i<9:
        i=2
    elif 9<=i<12:
        i=3
    elif 12<=i<15:
        i=4
    elif 15<=i<18:
        i=5
    elif 18<=i<21:
        i=6
    elif 21<=i<24:
        i=7
    sub.append(i)

f_copy['사고시간_인덱스']=sub
#print(f_copy.shape)
```
